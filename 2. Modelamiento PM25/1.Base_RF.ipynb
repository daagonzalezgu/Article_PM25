```python
import pandas as pd
```

```python
df_grilla = pd.read_excel('data/Grilla km2 Bogota.xlsx',sheet_name='Grilla 2km2')
df_grilla
```

```python
df_1 = df_grilla[df_grilla.Estacion.notnull()]
df_1.drop_duplicates(inplace=True)
df_1
```

```python
df_2 = df_grilla[df_grilla.Estacion.isnull()]
df_2.drop_duplicates(inplace=True)
df_2
```

```python
df = pd.concat([df_1,df_2])
df.drop_duplicates(keep='first',inplace=True)
df.reset_index(drop=True, inplace =True)
df
```

```python
###Agregar valores
df_datos = pd.read_excel('data/pm25_promedio.xlsx')
df_datos
```

```python
###Completitud
df_null = df_datos.isnull().sum().reset_index()
df_null.rename(columns={'index':'columna',0:'nulos'},inplace = True)
df_nonull= df_datos.notnull().sum().reset_index()
df_nonull.rename(columns={'index':'columna',0:'no_nulos'},inplace = True)
df_completitud = pd.merge(df_nonull,df_null,on='columna')
df_completitud['completitud'] = (1-df_completitud.nulos/(df_completitud.no_nulos +df_completitud.nulos))
df_completitud.sort_values(by='completitud',ascending = True)
```

```python
df_completitud1 = df_completitud[df_completitud['completitud'] >= 2.0/3.0].copy()
variables =  df_completitud1.columna.to_list()

variables
```

```python
df_datos = df_datos[variables].copy()
df_datos.rename(columns={'site':'Estacion'},inplace =True)
```

```python
#df_datos
```

```python
df_fechas = df_datos.groupby(["year","month"]).mean("pm25").reset_index()
df_fechas = df_fechas[["year","month"]]
df_fechas
```

```python
df_base00 = pd.merge(df,df_fechas,how='cross')
df_base00
```

```python
df_base0 = pd.merge(df_base00,df_datos,how='left',on= ["year","month","Estacion"])
df_base0.rename(columns={'i2':'y','j2':'x'},inplace =True)
df_base0
```

```python
##localidades
df_localidades = pd.read_excel('data/Localidades.xlsx')
df_localidades = df_localidades[["Código Localidad","Localidad" ]]
df_localidades.rename(columns = {"Código Localidad": "Codigo localidad"},inplace = True)
#df_localidades
```

```python
df_base01 = pd.merge(df_base0,df_localidades, how ='left', on = "Localidad")
df_base01
```

```python
###Numero de muertes por causas cardio-respiratorias
ruta_f1 = "../1. Identificacion muertes/base defunciones.csv" 
df_defunciones = pd.read_csv(ruta_f1,sep = ';',encoding='latin-1')
df_defunciones
```

```python
df_muertes = df_defunciones.groupby(["ANO","MES","Código localidad"]).count().reset_index()
df_muertes = df_muertes[["ANO","MES","Código localidad","Sexo"]]
df_muertes.rename(columns={"Código localidad":"Codigo localidad","Sexo":"MuertesxPM25",
                          "ANO":"year","MES":"month"},inplace =True)
df_muertes
```

```python
df_base02 = pd.merge(df_base01,df_muertes, how ='left', on = ["year","month","Codigo localidad"])
df_base02
```

```python
###Agregar poblacion
df_poblacion = pd.read_csv("data/osb_demografia-poblacion19072022.csv",sep = ';',encoding='latin-1')
df_poblacion.rename(columns ={"A¤o":"year","Cod_Loc":"Codigo localidad","Poblaci¢n": "Poblacion"},inplace = True)
df_pob = df_poblacion.groupby(["year","Codigo localidad"]).sum().reset_index()
df_pob = df_pob[["year","Codigo localidad","Poblacion"]]
df_pob
```

```python
##Final
df_base03 = pd.merge(df_base02,df_pob, how ='left', on = ["year","Codigo localidad"])
df_base03
```

```python
df_base03.to_csv('salidas/base_RF.csv')
```
