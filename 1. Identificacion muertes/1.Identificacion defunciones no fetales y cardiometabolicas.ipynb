```python
### El objetivo de este codigo es identificar las muertes en Bogota durante 2005 y 2020 por localidad,
###sexo y grupo etario, su causa y la identificación de las principales variables demograficas
```

```python
import pandas as pd
import os
```

```python
###Identificar lista de datos
files = os.listdir(os.getcwd()+'\\data')
```

```python
###Cargar lista de muertes por causas cardiometabolicas
ruta_causas = os.getcwd()+'\\parameters\\CausasCardioMetabolicas.xlsx'
df_causas = pd.read_excel(ruta_causas)
#df_causas
```

```python
#CARGAR HOMOLOGACION GRUPO ETARIO
ruta_edad = os.getcwd()+'\\parameters\\Homologación_Grupo_Edad.xlsx'
df_grupo_etario = pd.read_excel(ruta_edad)
df_grupo_etario.columns
```

```python
###CREAR DATAFRAME PARA AGRUPAR TODAS LOS REGISTROS DE DEFUNCIONES DEL DANE
df_DANE = pd.DataFrame(columns = ['COD_DPTO', 'COD_MUNIC', 'A_DEFUN', 'SIT_DEFUN', 'OTRSITIODE',
       'TIPO_DEFUN', 'ANO', 'MES', 'HORA', 'MINUTOS', 'SEXO', 'EST_CIVIL',
       'GRU_ED1', 'GRU_ED2', 'NIVEL_EDU', 'ULTCURFAL', 'MUERTEPORO',
       'SIMUERTEPO', 'OCUPACION', 'IDPERTET', 'CODPRES', 'CODPTORE',
       'CODMUNRE', 'AREA_RES', 'SEG_SOCIAL', 'IDADMISALUD', 'P_PMAN_IRIS',
       'CONS_EXP', 'MU_PARTO', 'T_PARTO', 'TIPO_EMB', 'T_GES',
       'T_GES_AGRU_CIE', 'PESO_NAC', 'EDAD_MADRE', 'N_HIJOSV', 'N_HIJOSM',
       'EST_CIVM', 'NIV_EDUM', 'ULTCURMAD', 'EMB_FAL', 'EMB_SEM', 'EMB_MES',
       'CODOCUR', 'CODMUNOC', 'C_MUERTE', 'C_MUERTEB', 'C_MUERTEC',
       'C_MUERTED', 'C_MUERTEE', 'ASIS_MED', 'CAUSA_MULT', 'C_BAS1',
       'CAUSA_667', 'IDPROFCER', 'CAU_HOMOL'])
```

```python
###AGRUPAR Y FILTRAR LAS MUERTES PARA EL MODELO
for file in files:
    ruta = 'data/'+ file
    df0 = pd.read_csv(ruta,sep=";",encoding='latin-1')
    ##SELECCIONAR DATOS DE BOGOTA (11 en el  codigo de departamento)
    df01 = df0[df0['COD_DPTO']==11].reset_index(drop=True)
    
    #SELECCIONAR LAS MUERTES EN BOGOTA DE HABITANTES DE BOGOTA
    df02 = df01[df01['CODPTORE']==11].reset_index(drop=True)
    
    ##SELECCIONAR MUERTES PARA PERSONAS CON 20 ANOS O MAS
    df_data00 = df02[df02['GRU_ED1'] >= 12].reset_index(drop=True)
    
    ###SELECCIONAR MUERTES CON CAUSAS CARDIOMETABOLICAS
    df_data0 = pd.merge(df_data00,df_causas, how = 'left', on = 'C_BAS1')
    df = df_data0[df_data0['Cardio-metabolicas']==1]
        
    ###CONCATENAR
    df_DANE = pd.concat([df_DANE,df], ignore_index=True)
```

```python
##CRUZAR CON HOMOLOGACION DEL GRUPO ETARIO
df_DANE.rename(columns={'GRU_ED1':'Grupo_Edad_DANE','SEXO':'GENERO'},inplace =True)
df_DANE0 = pd.merge(df_DANE,df_grupo_etario,how='left',on='Grupo_Edad_DANE')
df_DANE = df_DANE0[['COD_DPTO', 'COD_MUNIC', 'A_DEFUN', 'SIT_DEFUN', 'OTRSITIODE',
       'TIPO_DEFUN', 'ANO', 'MES', 'HORA', 'MINUTOS', 'GENERO', 'EST_CIVIL',
       'Grupo_Edad_DANE', 'GRU_ED2', 'NIVEL_EDU', 'ULTCURFAL', 'MUERTEPORO',
       'SIMUERTEPO', 'OCUPACION', 'IDPERTET', 'CODPRES', 'CODPTORE',
       'CODMUNRE', 'AREA_RES', 'SEG_SOCIAL', 'IDADMISALUD', 'P_PMAN_IRIS',
       'CONS_EXP', 'MU_PARTO', 'T_PARTO', 'TIPO_EMB', 'T_GES', 'PESO_NAC',
       'EDAD_MADRE', 'N_HIJOSV', 'N_HIJOSM', 'EST_CIVM', 'NIV_EDUM',
       'ULTCURMAD', 'EMB_FAL', 'EMB_SEM', 'EMB_MES', 'CODOCUR', 'CODMUNOC',
       'ASIS_MED', 'CAUSA_MULT', 'C_BAS1', 'CAUSA_667', 'IDPROFCER',
       'CAU_HOMOL', 'Grupo_Edad']]
df_DANE['NO_DANE'] = df_DANE.index
```

```python
##MUERTES POR LOCALIDADES
ruta_localidad = os.getcwd()+'\\parameters\\osb_demografia-causasmortalidad.csv'
df_localidad = pd.read_csv(ruta_localidad,sep=';',encoding='latin-1')
df_localidad.drop(['MORTALIDAD_DESCRIPCION_105',
       'MORTALIDAD_DESCRIPCION_667', 'MORTALIDAD_CONDICION_CRONICA'], axis = 1, inplace = True)
df_localidad.rename(columns = {'ANIO': 'ANO', 'EDAD_GRUPO':'Grupo_Edad_Causa' ,
                                     'MORTALIDAD_CODIGO_CIE10':'C_BAS1',
                              'SEXO':'Sexo', 'LOCALIDAD_CODIGO' :'Código localidad',
                              'LOCALIDAD_NOMBRE': 'Localidad', 
                               'SEGURIDAD_SOCIAL_REGIMEN':'Régimen de seguridad social',
                              'MORTALIDAD_DESCRIPCION_CIE10':'Descripción causa básica de muerte'},inplace =True)
##ESTANDARIZAR GENERO
df_localidad['GENERO'] = df_localidad.Sexo.apply(lambda x: 1 if x == 'Masculino' else
                                                                    2 if x == 'Femenino' else 3)
##HOMOLOGAR REGIMEN DE SEGURIDAD SOCIAL
df_localidad['SEG_SOCIAL'] = df_localidad['Régimen de seguridad social'].apply(lambda x: 
                                   1 if x == "CONTRIBUTIVO" else
                                   2 if x == "SUBSIDIADO" else 
                                   3 if x == "EXCEPCION" else
                                   4 if x == "ESPECIAL" else
                                   5 if x == "NO ASEGURADO" else 6)
##CRUZAR CON HOMOLOGACION GRUPO ETARIO
df_muertes_bogota00 = pd.merge(df_localidad,df_grupo_etario,how='left',on='Grupo_Edad_Causa')
##SELECCIONAR MUERTES PARA PERSONAS CON 20 ANOS O MAS
df_muertes_bogota =  df_muertes_bogota00[df_muertes_bogota00['Grupo_Edad_DANE'] >= 12].reset_index(drop=True)

##QUITAR COLUMNAS NO NECESARIAS
df_muertes_bogota.drop(['Grupo_Edad_Causa','Grupo_Edad_poblacion', 
                        'Grupo_Edad_DANE'], axis = 1, inplace = True)
##FILTAR MUERTES DESDE EL 2008
df_muertes_bogota =df_muertes_bogota[df_muertes_bogota['ANO']>=2008].reset_index(drop=True)

###SELECCIONAR MUERTES CON CAUSAS CARDIOMETABOLICAS
df_muertes_bogota0 = pd.merge(df_muertes_bogota,df_causas, how = 'left', on = 'C_BAS1')
df_muertes_bogota = df_muertes_bogota0[df_muertes_bogota0['Cardio-metabolicas']==1]




##INDEXAR PARA EVITAR DUPLICIDAD Y QUITAR REGISTROS 
df_muertes_bogota['NO'] = df_muertes_bogota.index
df_muertes_bogota
```

```python
# GENERAR BASE DE DEFUNCIONES: CRUZAR POR ANO, GENERO, GRUPO ETARIO,LOCALIDAD Y CAUSA DESDE MUERTES BOGOTA ---> DANE
df1= pd.merge(df_muertes_bogota,df_DANE, how = 'left',
              on = ['ANO','GENERO','Grupo_Edad','C_BAS1','SEG_SOCIAL'])
df1.drop_duplicates(subset=['NO'],keep = 'first',inplace=True)
df1.fillna(0,inplace =True)
df1['Código localidad'] = df1['Código localidad'].astype('int64')
df1 = df1[df1['Código localidad'] != 0]
df1 = df1[df1['MES'] != 0]
df1 = df1[df1['Grupo_Edad_DANE'] != 0]
```
