```python
import pandas as pd
import os
import random
```

```python
###LEER SALIDA DE LA PARTE 1: IDENTIFICACION DEFUNCIONES NO FETALES Y CARDIOMETABOLICAS
ruta_f1 = "../1. Identificacion muertes/base defunciones.csv" 
df_mortalidad = pd.read_csv(ruta_f1,sep = ';',encoding='latin-1')
df_mortalidad.rename(columns = {'Código localidad':'Codigo localidad'},inplace = True)
#df_f1
```

```python
##LEER SALIDA DE LA PARTE 2: MODELO DE ESTIMACION MENSUAL DE PM25 POR LOCALIDAD
ruta_f2 = "../2. Modelamiento PM25/salidas/pm25_pred_anual.xlsx" 
df_PM = pd.read_excel(ruta_f2)
df_PM.rename(columns = {'year':'ANO'},inplace = True)
#df_PM = df_PM[['ANO', 'MES', 'site', 'PM_25_model', 'std']]
df_PM.head()
```

```python
##CRUZAR POR CAMPOS COMPLETOS
df_base00 = pd.merge(df_mortalidad,df_PM,how = 'inner', on = ['ANO','Codigo localidad'])
df_base00
```

```python
##GUARDAR
df_base00.to_excel('salidas/base_completa.xlsx',index=False)
```

```python
df_R = df_base00[['Codigo localidad','Localidad', 'ANO','MES','Grupo_Edad_DANE','GENERO',
                'pm25_pred','pm25_pred_prom','SEG_SOCIAL', 'EST_CIVIL', 'NIVEL_EDU','IDPERTET']]
###ESTATUS = 1 CORRESPONDE A QUE EL REGISTRO CORRESPONDE A UNA MUERTE
###NO SE TIENEN REGISTROS SESGADOS
df_R['estatus'] = 1
###COMPARAR LAS DEFUNCIONES EN HOMBRES RESPECTO A LAS MUJERES
df_R['sexo'] = df_R.GENERO.apply(lambda x: 0 if x == 1 else 1)

###COMPARAR LAS DEFUNCIONES ENTRE LOS NO CASADOS Y LOS CASADOS
df_R['casado'] = df_R.EST_CIVIL.apply(lambda x: 0 if x == 5 else 1)


####COMPARAR LAS DEFUNCIONES ENTRE LOS NO ASEGURADOS RESPECTO A LOS ASEGURADOS
#df_R['no_asegurado'] = df_R.SEG_SOCIAL.apply(lambda x: 0 if x <5 else 1)

###COMPARAR LAS DEFUNCIONES ENTRE LOS QUE TIENEN BASICA Y NO 
#df_R['mas_basica'] = df_R.NIVEL_EDU.apply(lambda x: 0 if x <4 else 1)


###
df_R['PM25_base'] = df_R.pm25_pred.apply(lambda x: 0 if x <11 else 1)
#------
###TOMAR EL TIEMPO EN EL ESTUDIO A PARTIR DEL 01/2008
#df_R['tiempo'] = (df_R['ANO']-2008)*12+df_R['MES']
###TIEMPO ALTERNO - MES DEL ANO
#df_R['tiempo'] = df_R['MES']
df_R['tiempo'] = df_R['ANO'] -2007
##-----
###ASIGNAR INDICES DE FORMA ALEATORIA 
s = []
for x in range(0,len(df_R)):
    s.append(random.randint(0,len(df_R)))
df_R['orden'] = s
df_R.sort_values('orden',inplace = True,ignore_index=True)
df_R
```

```python
###CREAR CATEGORIAS PARA EDADES
for edad in range(18,28):
    df_R['edad_'+str(edad)] = df_R.Grupo_Edad_DANE.apply(lambda x: 1 if x >= edad else 0)
```

```python
df_R.rename(columns= {'Codigo localidad':'localidad','Grupo_Edad_DANE':'edad'},inplace=True)
```

```python
###CREAR CATEGORIAS PARA SEGURIDAD SOCIAL
for i in df_R.SEG_SOCIAL.unique():
    df_R['Seg_social_'+str(i)] = df_R.SEG_SOCIAL.apply(lambda x: 1 if x == i else 0)
```

```python
##CREAR CATEGORIAS PARA NIVEL EDUCATIVO
for i in df_R.NIVEL_EDU.unique():
    df_R['Edu_'+str(int(i))] = df_R.NIVEL_EDU.apply(lambda x: 1 if x == i else 0)
```

```python
##CREAR CATEGORIAS PARA ETNIA
for i in df_R.IDPERTET.unique():
    df_R['Grupo_'+str(int(i))] = df_R.IDPERTET.apply(lambda x: 1 if x == i else 0)
```

```python
####CREAR CATEGORIAS PARA ESTADO CIVIL
for i in df_R.EST_CIVIL.unique():
    df_R['EstCiv_'+str(int(i))] = df_R.EST_CIVIL.apply(lambda x: 1 if x == i else 0)
```

```python
df_R.columns
```

```python
df_R['Posgrado'] = 1-(df_R['Edu_11']+df_R['Edu_12'])
df_R['Maestria'] = 1-df_R['Edu_11']
df_R['Etnia'] = df_R['Grupo_2']+df_R['Grupo_4']
df_R['Etnia2'] = df_R['Grupo_2']+df_R['Grupo_4']+df_R['Grupo_3']
```

```python
df_R = df_R[['localidad', 'ANO', 'MES', 'edad', 'sexo', 'pm25_pred','pm25_pred_prom',
        'estatus', 'casado', 'PM25_base', 'tiempo', 'orden',
       'edad_18', 'edad_19', 'edad_20', 'edad_21', 'edad_22', 'edad_23',
       'edad_24', 'edad_25', 'edad_26', 'edad_27','Seg_social_1',
       'Seg_social_2', 'Seg_social_5', 'Seg_social_3', 'Seg_social_4',
        'Edu_8', 'Edu_2', 'Edu_99', 'Edu_10', 'Edu_9',
       'Edu_3', 'Edu_4', 'Edu_1', 'Edu_13', 'Edu_7', 'Edu_5', 'Edu_11',
       'Edu_6', 'Edu_12', 'Grupo_6', 'Grupo_9', 'Grupo_5', 'Grupo_1',
       'Grupo_2', 'Grupo_3', 'Grupo_4','Posgrado','Maestria','Etnia','Etnia2','EstCiv_1'
            ,'EstCiv_2','EstCiv_3','EstCiv_4','EstCiv_5','EstCiv_6']]
df_R
```

```python
df_R['tiempo2'] = (df_R.ANO-2008)*12+df_R.MES
df_R['tiempo3'] = df_R['tiempo']+df_R['MES']/13.0
```

```python
df_R['tiempo4'] = df_R['tiempo3'] + abs(df_R['sexo']-1)/26.0
```

```python
df_R['NoViudo'] = 1-df_R['EstCiv_4']
#df_R['NoCasado']
```

```python
df_R['NoSubsidiado'] = 1 - df_R['Seg_social_2']
```
